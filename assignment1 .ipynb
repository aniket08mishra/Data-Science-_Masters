```python
# Q1. Example of when to use a for loop and a while loop

# For loop: When you know the exact number of iterations
numbers = [1, 2, 3, 4, 5]
print("Using for loop:")
for num in numbers:
    print(num)

# While loop: When the number of iterations is not known beforehand
count = 1
print("\nUsing while loop:")
while count <= 5:
    print(count)
    count += 1

# Q2. Sum and product of the first 10 natural numbers using for and while loops

# Using for loop
sum_for = 0
product_for = 1
for i in range(1, 11):
    sum_for += i
    product_for *= i
print("\nSum using for loop:", sum_for)
print("Product using for loop:", product_for)

# Using while loop
sum_while = 0
product_while = 1
i = 1
while i <= 10:
    sum_while += i
    product_while *= i
    i += 1
print("\nSum using while loop:", sum_while)
print("Product using while loop:", product_while)

# Q3. Compute the electricity bill for a household

def calculate_electricity_bill(units):
    if units <= 100:
        bill = units * 4.5
    elif units <= 200:
        bill = (100 * 4.5) + ((units - 100) * 6)
    elif units <= 300:
        bill = (100 * 4.5) + (100 * 6) + ((units - 200) * 10)
    else:
        bill = (100 * 4.5) + (100 * 6) + (100 * 10) + ((units - 300) * 20)
    return bill

# Test case
units = 310
print("\nTotal electricity bill:", calculate_electricity_bill(units))  # Output should be 2250

# Q4. Calculate the cube of each number from 1 to 100 and append numbers whose cubes are divisible by 4 or 5 to a list

# Using for loop
result_for = []
for i in range(1, 101):
    cube = i ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        result_for.append(i)
print("\nNumbers whose cubes are divisible by 4 or 5 using for loop:", result_for)

# Using while loop
result_while = []
i = 1
while i <= 100:
    cube = i ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        result_while.append(i)
    i += 1
print("Numbers whose cubes are divisible by 4 or 5 using while loop:", result_while)

# Q5. Program to filter and count vowels in a given string

string = "I want to become a data scientist"
vowels = 'aeiouAEIOU'
vowel_count = {v: 0 for v in vowels}

for char in string:
    if char in vowels:
        vowel_count[char] += 1

# Display the count of each vowel
print("\nVowel counts in the given string:")
for vowel, count in vowel_count.items():
    if count > 0:
        print(f"{vowel}: {count}")
```
